```python

features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]

numeric_cols = ['floors', 'waterfront', 'lat', 'sqft_basement', 'bathrooms', 'sqft_living15', 'sqft_above', 'grade', 'sqft_living']
categorical_cols = ['bedrooms', 'view']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numeric_cols),
        ('cat', OneHotEncoder(), categorical_cols)
    ])

pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('polynomial', PolynomialFeatures(include_bias=False)),
    ('model', LinearRegression())
])

X = df[features]
y = df['price']

pipe.fit(X, y)

y_pred = pipe.predict(X)
rsquared = r2_score(y, y_pred)
print("R-squared:", rsquared)
```

```python
# Handling missing values
df = df.dropna()  # Dropping rows with any NaN values, you might choose a different strategy based on your data

# Features and target variable
x1 = df[['sqft_living']]
y1 = df['price']

# Splitting the data into training and testing sets for simple linear regression
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.15, random_state=1)

# Simple Linear Regression
lm1 = LinearRegression()
lm1.fit(x1_train, y1_train)
y1_predict = lm1.predict(x1_test)
r_squared_simple = r2_score(y1_test, y1_predict)
print("R-Squared for Simple Linear Regression:", r_squared_simple)

# Now, for Ridge Regression using the previously defined features and target variables
features = ["sqft_living"]
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)

# Create Ridge regression object
ridge_reg = Ridge(alpha=0.1)  # Set the regularization parameter (alpha) to 0.1

# Fit the model using the training data
ridge_reg.fit(x_train, y_train)

# Predict on the test set
y_pred = ridge_reg.predict(x_test)

# Calculate R^2 score
r2_ridge = r2_score(y_test, y_pred)
print("R^2 score using Ridge Regression:", r2_ridge)
```
