```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pandas as pd


# Extract categorical columns
categorical_columns = X.select_dtypes(include=['object']).columns

# Instantiate LabelEncoder
label_encoder = LabelEncoder()

# Encode the target variable for the entire dataset
y_encoded = label_encoder.fit_transform(y)

# One-hot encode categorical features
X_encoded = pd.get_dummies(X, columns=categorical_columns, drop_first=True)

# Split the dataset into training and testing sets
X_train, X_test, y_train_encoded, y_test_encoded = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# Initialize a Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)

# Train the model on the training set
rf_model.fit(X_train, y_train_encoded)

# Make predictions on the test set
y_pred_encoded = rf_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test_encoded, y_pred_encoded)
print(f'Random Forest Accuracy: {accuracy:.2f}')

# Calculate mean squared error
rf_mse = mean_squared_error(y_test_encoded, y_pred_encoded)
print(f'Random Forest Mean Squared Error: {rf_mse:.4f}')

# Calculate mean absolute error
rf_mae = mean_absolute_error(y_test_encoded, y_pred_encoded)
print(f'Random Forest Mean Absolute Error: {rf_mae:.4f}')

# Calculate R2 score error
rf_r2score = mean_absolute_error(y_test_encoded, y_pred_encoded)
print(f'Random Forest R2 Score: {rf_r2score:.4f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Assuming 'ufc_df' is your DataFrame
# Replace this with your actual data loading code

# Extract relevant features and target variable
features = ['Fighter_2_KD', 'Fighter_1_TD', 'Fighter_2_TD', 'Fighter_1_SUB', 'Fighter_2_SUB']
target = 'Winner'

X = ufc_df[features]
y = ufc_df[target]

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Initialize models
svm_model = SVC(random_state=42)
logistic_model = LogisticRegression(random_state=42)
rf_model = RandomForestClassifier(random_state=42)

# Train models
svm_model.fit(X_train, y_train)
logistic_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)

# Make predictions
svm_pred = svm_model.predict(X_test)
logistic_pred = logistic_model.predict(X_test)
rf_pred = rf_model.predict(X_test)

# Calculate accuracy for each model
svm_accuracy = accuracy_score(y_test, svm_pred)
logistic_accuracy = accuracy_score(y_test, logistic_pred)
rf_accuracy = accuracy_score(y_test, rf_pred)

# Create a DataFrame for easier plotting
metrics_df = pd.DataFrame({
    'Model': ['SVM', 'Logistic Regression', 'Random Forest'],
    'Accuracy': [svm_accuracy, logistic_accuracy, rf_accuracy]
})

# Create a bar plot
plt.figure(figsize=(5, 6))
sns.barplot(x='Model', y='Accuracy', data=metrics_df, palette='viridis')
plt.title('Model Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set the y-axis limit to better visualize differences
plt.show()
```
